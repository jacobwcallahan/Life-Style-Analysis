```python
corr = data.corr(numeric_only=True)
np.fill_diagonal(corr.values, np.nan)
top_corr = corr.abs().unstack().sort_values(ascending=False).head(30)
top_corr.drop_duplicates(inplace=True)
top_corr
```

The biggest thing to note from the correlation matrix is the variables `BMI` and `BMI_calc` are perfectly correlated. We will drop `BMI_calc` so it doens't disrupt our analysis. 

As well, `Fats`, `Proteins`, `Carbs` and `cal_from_macros` are also all pretty closely correlated. `Fats`, `Proteins`, and `Carbs` will be removed to maintain quality data for analysis. 

Lastly, `pct_maxHR` will be dropped as it is closely tied with `pct_HRR`. 

There may be other values dropped later but currently it will just be these.

```python
data = data.drop(columns=['BMI_calc', 'Fats', 'Proteins', 'Carbs', 'pct_maxHR'])
```
